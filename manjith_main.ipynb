```python
import pandas as pd
import numpy as np
# using train data set to extrat unique items of a row
dfbuff = pd.read_excel('C:/Users/manji/OneDrive/Desktop/Main/New folder/OneDrive_1_1-6-2023/housing_data/train.xlsx')
furnishing={'Unfurnished':0,'Semi-Furnished':1,'Furnished':2}
tennants={'Bachelors':0,'Family':1,'Bachelors/Family':2}
bedrooms={}
buff=dfbuff['Bedrooms'].unique()
# creating a dictionary for labeling
for i in range(0,len(buff)):
    bedrooms[buff[i]]=i+1
print(bedrooms)
# defing a function to remove null values every time labeling is done
def removeNull(df1):
        NA_Count = pd.DataFrame({'Sum of NA':df1.isnull().sum()}).sort_values(by=['Sum of NA'],ascending=[0])
        NA_Count['Percentage'] = NA_Count['Sum of NA']/df1.shape[1]
        sum(NA_Count['Percentage'])
        df1=df1.dropna()
        NA_Count = pd.DataFrame({'Sum of NA':df1.isnull().sum()}).sort_values(by=['Sum of NA'],ascending=[0])
        NA_Count['Percentage'] = NA_Count['Sum of NA']/df1.shape[1]
        sum(NA_Count['Percentage'])
        return df1
# defing a function for labeling, string-extraction, removing null values, to map and add new colunms
def preprocess (address):
    locations=["Nampally","Shamshabad","Hayathnagar","Vanasthalipuram","Dilshuknagar","Dilsukh Nagar","LB Nagar","Nagole","Bowenpally","Kompally","Alwal","Sainikpuri","ECIL","Shamirpet","Miyapur","Nizampet","Patancheru","Rajendranagar","Pocharam","Boduppal","Ghatkesar","Banjara Hills","Hitech City","Jubilee Hills","Madhapur","Gachibowli","Kondapur","Manikonda","Narsingi","Golkonda","Mehdipatnam","Attapur","KPHB","Ameerpet","Uppal","Tarnaka","Habsiguda","Amberpet","Punjagutta","Nallakunta","Narayanguda","Masab Tank","Khairatabad","Himayatnagar","Begumpet","Secunderabad","Marredpally","Nacharam","Mallapur","Oldcity","Malakpet"]
    pincodes=["500095","500079","500079","500079","500068","500068","500068","500068","500067","500067","500067","500062","500062","500062","500050","500050","500050","500048","500039","500039","500039","500033","500033","500033","500033","500032","500032","500032","500032","500028","500028","500028","500018","500018","500007","500007","500007","500007","500004","500004","500004","500004","500004","500004","500004","500003","500003","500003","500003","500001","500001"]
    locpin={}
    for i in range(0,len(locations)):
        locpin[locations[i]]=int(pincodes[i])
    
    df = pd.read_excel(address)
    df=removeNull(df)
    Pincodes=[]
    for i in range(0,len(df)):
        try:
            Pincodes.append(locpin[df['Location'].values[i]])
        except:
            Pincodes.append(None)
    
    df.loc[:,'PinCodes'] = Pincodes
    df=removeNull(df)
    for i in range(0,len(df)):
        try:
            df['Furnishing'].values[i]=furnishing[df['Furnishing'].values[i]]
        except:
            df['Furnishing'].values[i]=None   
    df=removeNull(df)
    for i in range(0,len(df)):
        try:
            df['Tennants'].values[i]=tennants[df['Tennants'].values[i]]
        except:
            df['Tennants'].values[i]=None   
    df=removeNull(df)
    for i in range(0,len(df)):
        try:
            df['Area'].values[i]=int(df['Area'].values[i][:-4].replace(",",""))
        except:
            df['Area'].values[i]=None
    df=removeNull(df)
    for i in range(0,len(df)):
        try:
            df['Bedrooms'].values[i]=bedrooms[df['Bedrooms'].values[i]]
        except:
            df['Bedrooms'].values[i]=None
    df=removeNull(df)
    for i in range(0,len(df)):
        try:
            if(type(df['Price'].values[i])==type('str')):
                if(','in df['Price'].values[i]):
                    df['Price'].values[i]=int(df['Price'].values[i].replace(",",""))
                else:
                    df['Price'].values[i]=int(df['Price'].values[i])

        except:
            df['Price'].values[i]=None
    df=removeNull(df)
    df=df.drop(columns=['Location', 'Locality','Facing'])
    return df
```

```python
# calling preprocess function with train dataset address
dfToTrain=preprocess('C:/Users/manji/OneDrive/Desktop/Main/New folder/OneDrive_1_1-6-2023/housing_data/train.xlsx')
print(len(dfToTrain))
dfToTrain.head()
```

```python
# finding corelation between different variables
dfToTrain.corr(numeric_only=False)
```

```python
# consedering only columns with positive corelation
# by trail and error method it is found that 2 degree multivariat polynomial regression is best fit
from sklearn.linear_model import LinearRegression  
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline
polyreg=make_pipeline(PolynomialFeatures(2),LinearRegression())
# definig a function to form a regression model with selected columns for train dataset
def train():
    x=dfToTrain[['Bathrooms',"Furnishing",'Area']]
    y=dfToTrain['Price']
    polyreg.fit(x,y)

```

```python
# calling preprocess function with test dataset address
dfToTest=preprocess('C:/Users/manji/OneDrive/Desktop/Main/New folder/OneDrive_1_1-6-2023/housing_data/test.xlsx')
print(len(dfToTest))
dfToTest.head()
```

```python
# defining a function test to compare the results of predection to actual values
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
def test():
    x_test=dfToTest[['Bathrooms',"Furnishing",'Area']]
    print(x_test)
    y_test=dfToTest['Price']
    ypred=polyreg.predict(x_test)
    print(ypred)
    score=r2_score(y_test,ypred)
    print("r2 socre is" ,score)
    print("mean_sqrd_error is=",mean_squared_error(y_test,ypred))
    print("root_mean_squared error of is=",np.sqrt(mean_squared_error(y_test,ypred)))
```

```python
# by trail and error it is found out that there are some outlyers in both train and test datasets which are being removed
Aq1 = dfToTrain["Area"].quantile(0.99)
Aq2 = dfToTrain["Area"].quantile(0.01)
Pq1=dfToTrain["Price"].quantile(0.99)
Pq2 = dfToTrain["Price"].quantile(0.01)
dfToTrain = dfToTrain[(dfToTrain["Area"] < Aq1) & (dfToTrain["Area"] > Aq2)]
dfToTrain = dfToTrain[(dfToTrain["Price"] < Pq1) & (dfToTrain["Price"] > Pq2)]
print(len(dfToTrain))
Aq1 = dfToTest["Area"].quantile(0.99)
Aq2 = dfToTest["Area"].quantile(0.01)
Pq1=dfToTest["Price"].quantile(0.99)
Pq2 = dfToTest["Price"].quantile(0.01)
dfToTest = dfToTest[(dfToTest["Area"] < Aq1) & (dfToTest["Area"] > Aq2)]
dfToTest = dfToTest[(dfToTest["Price"] < Pq1) & (dfToTest["Price"] > Pq2)]
print(len(dfToTest))

```

```python
# calling train and test functions 
train()
test()
```
